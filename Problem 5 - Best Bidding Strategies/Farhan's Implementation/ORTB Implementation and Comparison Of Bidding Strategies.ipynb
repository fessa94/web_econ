## Bidding

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pctrval = pd.read_csv('pCTRval.csv')
df_val = pd.read_csv('data/validation.csv')
average_ctr = df_val.click.sum() / df_val.bidid.count()
pctrval_list = list(pctrval['0'])
```

## Linear Bidding

```python
def linear_bidding(pctrval_list):
    bids = []
    
    base_bid = 210
    
    for pctr in pctrval_list:
        bids.append(base_bid * pctr / average_ctr)

    return bids
```

```python
df_ortb = pd.DataFrame()
df_ortb['bids'] = linear_bidding(pctrval_list)
df_valortb = pd.concat([df_val, df_ortb], axis=1)
df_valortb
```

```python
data3 = pd.DataFrame(columns=['clicks','imps','spent','CTR','CPC','CPM'])
lists = {'clicks':[], 'imps':[], 'spent':[], 'ctr':[], 'cpc':[], 'cpm':[]}

budget = 6250000.0
imps = 0
clicks = 0
spent = 0
for i in range(0,len(df_valortb)):
    payprice = df_valortb.payprice[i]
    bid = df_valortb.bids[i]
    if budget < payprice:
        break
    if bid >= payprice:
        imps = imps + 1
        clicks = clicks + df_valortb.click[i]
        budget = budget - payprice
        spent = spent + payprice
spent = spent / 1000
ctr = ((clicks / imps) * 100).round(4).astype(str)
cpm = ((spent / imps) * 1000).round(4).astype(str)
cpc = (spent / clicks).round(4).astype(str)
lists['clicks'].append(clicks)
lists['imps'].append(imps)
lists['spent'].append(spent)
lists['ctr'].append(ctr)
lists['cpc'].append(cpc)
lists['cpm'].append(cpm)

data3.clicks = lists['clicks']
data3.imps = lists['imps']
data3.spent = lists['spent']
data3.CTR = lists['ctr']
data3.CPC = lists['cpc']
data3.CPM = lists['cpm']
data3
```

## ORTB - Best Bidding Strategy

```python
def ortb_bidding(pctrval_list):
    bids = []
    c = 28
    l = 5e-07
    
    for pctr in pctrval_list:
        bids.append(np.sqrt(((c/l)*pctr)+(c*c)-c))
    
    return bids
```

```python
df_ortb = pd.DataFrame()
df_ortb['bids'] = ortb_bidding(pctrval_list)
df_valortb = pd.concat([df_val, df_ortb], axis=1)
df_valortb
```
